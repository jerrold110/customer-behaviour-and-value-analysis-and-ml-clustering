```python
import pandas as pd
import datetime as dt
```

```python
online = pd.read_csv('online_orders.csv')
print(online.shape)
```

```python
online['InvoiceDate'] = pd.to_datetime(online['InvoiceDate'])
online.drop(labels=['Unnamed: 0'], axis=1, inplace=True)
online.head()
```

# Recency, Frequency, Monetary Value

* Today is one day after the last purchase

* Time period for analysis is last 12 months

### Recency (days since last purchase)
### Frequency (how many purchases in the last 12 months)
### Monetary value (How much spent in the last 12 months)

```python
purchase = online[['CustomerID', 'Quantity', 'UnitPrice', 'InvoiceDate']].copy()
purchase.head()
```

```python
purchase['CustomerID'].unique().shape
```

```python
pd.to_datetime('2011-12-30') + dt.timedelta(days=1)
```

```python
# Recency (days since last purchase)
# Frequency (how many purchases in the last 12 months)
# Monetary value (How much spent in the last 12 months)

today = purchase['InvoiceDate'].max() + dt.timedelta(days=1)
cutoff = purchase['InvoiceDate'].max() - dt.timedelta(days=365)
print(f'Today: {today}. 12M cutoff: {cutoff}')

purchase = purchase[(purchase['InvoiceDate']<today) & (purchase['InvoiceDate']>=cutoff)]
purchase['TotalSum'] = purchase['Quantity']*purchase['UnitPrice']
rfm = purchase.groupby(['CustomerID']).agg({
    'InvoiceDate':lambda x: (today-x.max()).days,
    'Quantity': 'count',
    'TotalSum': 'sum'
})

rfm.rename(columns={'InvoiceDate':'Recency',
                   'Quantity':'Frequency',
                   'TotalSum':'Monetary_value'}, inplace=True)
rfm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for abc in ['Recency', 'Frequency', 'Monetary_value']:
    plt.figure(figsize=(10,4))
    sns.histplot(data=rfm, x=abc, bins=50)
    plt.show()
```

# RFM segments

```python
r_labels = range(4, 0, -1)
r_quartiles = pd.qcut(rfm['Recency'], 4, labels=r_labels) # orders values in ascending order
rfm = rfm.assign(R=r_quartiles.values)
rfm
```

```python
f_labels = range(1,5,1)
m_labels = range(1,5,1)

f_quartiles = pd.qcut(rfm['Frequency'], 4, labels=f_labels)
m_quartiles = pd.qcut(rfm['Monetary_value'], 4, labels=m_labels)

rfm = rfm.assign(F=f_quartiles)
rfm = rfm.assign(M=m_quartiles)
rfm.head()
```

# Segments analysis

```python
def joinrfm(x):
    return str(x['R'])+str(x['F'])+str(x['M'])

rfm['Segment'] = rfm.apply(joinrfm, axis=1)
rfm['RFM_score'] = rfm[['R','F','M']].sum(axis=1)
rfm
```

```python
plt.figure(figsize=(10,30))
sns.countplot(data=rfm.sort_values('Segment', ascending=False), y='Segment').set_title('Count of segments')

plt.show()
```

```python
rfm.groupby('RFM_score').agg({
    'Recency':'mean',
    'Frequency':'mean',
    'Monetary_value':['mean','count']})\

.round(1)
```

```python
# For clustering segmentation
clustering_breadth = online[(online['InvoiceDate']<today) & (online['InvoiceDate']>=cutoff)]
clustering_breadth = clustering_breadth.groupby('CustomerID')[['StockCode']].nunique().rename(columns={'StockCode':'Breadth'})

clustering_data = rfm[['Recency', 'Frequency', 'Monetary_value']].join(clustering_breadth, how='left')
clustering_data.head()
```

```python
clustering_data.to_csv('data/clustering_datamart.csv')
```
